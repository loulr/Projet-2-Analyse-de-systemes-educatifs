### Secondary

#### Identification des variables les plus importantes

```python
# Filtrage auto sur les indicateurs intéressants
df_sec = dfSeries[(dfSeries['Topic'] == 'Secondary')]
a = dfSeries['Series Code'].apply(lambda x: x.split("."))

# Filtrage sur les indicateurs les moins déclinés
df = pd.DataFrame(df_sec[['Series Code', 'Indicator Name']])
df["b"] = df_sec['Series Code'].apply(lambda x: x.split("."))
df["a"] = df_sec['Series Code'].apply(lambda x: x.split(".")[0])
df["n"] = df['b'].apply(lambda x: len(x))

# Recherche sur .ENR pour enrollment
df = df[df['Series Code'].str.contains('\.ENR', regex=True)].sort_values("n")

# Recherche des codes topic uniques
a = list(set(df["a"]))
print("Sous-domaines qui nous intéressent :", a)

# Affichage des indicateurs retenus
b = df['Series Code'][:2].reset_index(drop=False)
print("Indicateurs retenus : ", b)
```

Pour le niveau secondaire, nous avons identifié deux variables utilisables : 
- "SE.SEC.ENRL" Total number of students enrolled at public and private secondary education institutions regardless of age.
- "SE.SEC.ENRR" Gross enrolment ratio, secondary, both sexes (%)

Dans la suite, nous allons regarder les taux de remplissage et le choix des pays et des années.

#### Base de données par thématique

```python
# Filtrage sur les pays et non régions
dfData = dfData[dfData['Country Name'].isin(countries)]

# Filtrage sur indicateur SE.SEC.ENRL
dfDataFiltered_SE_SEC_ENRL = dfData[(dfData['Indicator Code'] == b["Series Code"][0])]

# Filtrage sur indicateur SE.SEC.ENRL
dfDataFiltered_SE_SEC_ENRR = dfData[(dfData['Indicator Code'] == b["Series Code"][1])]
```

#### Evaluation de la qualité des données (taux de remplissage, duplication)

```python
# Taux de missing data sur les années sur ENRL
missing_annees = dfDataFiltered_SE_SEC_ENRL.isnull().mean()*100

# Taux de missing data sur les pays sur ENRL
dfDataFiltered_SE_SEC_ENRL.index = dfDataFiltered_SE_SEC_ENRL['Country Name']
missing_pays = dfDataFiltered_SE_SEC_ENRL.transpose().isnull().mean()*100


print("Taux de missing data :")
plt.subplot(1, 2, 1)
plt.hist(missing_pays, label='Missing Pays')
plt.hist(missing_annees, label='Missing Annees')
plt.legend(loc="upper left")
plt.title('Missing data Pays/Annees ENRL')
plt.xlabel('Taux de Missing Data %')
plt.ylabel('Répartition %')


# Taux de missing data sur les années sur ENRL
missing_annees = dfDataFiltered_SE_SEC_ENRR.isnull().mean()*100

# Taux de missing data sur les pays sur ENRL
dfDataFiltered_SE_SEC_ENRR.index = dfDataFiltered_SE_SEC_ENRR['Country Name']
missing_pays = dfDataFiltered_SE_SEC_ENRR.transpose().isnull().mean()*100

plt.subplot(1, 2, 2)
plt.hist(missing_pays, label='Missing Pays')
plt.hist(missing_annees, label='Missing Annees')
plt.legend(loc="upper left")
plt.title('Missing data Pays/Annees ENRR')
plt.xlabel('Taux de Missing Data %')
plt.ylabel('Répartition %')


plt.tight_layout()
```

**Observations**: On remarque que les deux variables ENRL (nombre d'étudiants inscrits) et ENRR (taux de scolarisation) ont sensiblement les mêmes remplissages de données. On peut donc utiliser l'une ou l'autre. Ici nous utiliserons ENRR afin d'avoir un taux relatif à la population directement.

#### Choix des variables années et pays sur lesquels porteront l'analyse

```python
# On enlève les pays de plus de 80% de missing data
pays_keep = missing_pays[missing_pays < 80]
dfDataFiltered_SE_SEC_ENRR = dfDataFiltered_SE_SEC_ENRR[dfDataFiltered_SE_SEC_ENRR['Country Name'].isin(pays_keep.index.tolist())]

# On enlève les années de plus de 50% de missing data
annees_keep = missing_annees[missing_annees < 50]
dfDataFiltered_SE_SEC_ENRR = dfDataFiltered_SE_SEC_ENRR[annees_keep.index.tolist()]
```

#### Statistiques descriptives et visualisation 

```python
# Création des colonnes moyenne, médiane, variance et nombre de NaN
dfDataFiltered_SE_SEC_ENRR['Mean'] = dfDataFiltered_SE_SEC_ENRR.mean(axis=1)
dfDataFiltered_SE_SEC_ENRR['Median'] = dfDataFiltered_SE_SEC_ENRR.median(
    axis=1)
dfDataFiltered_SE_SEC_ENRR['Std'] = dfDataFiltered_SE_SEC_ENRR.std(axis=1)
dfDataFiltered_SE_SEC_ENRR['Nb NaN'] = dfDataFiltered_SE_SEC_ENRL.isnull().sum(
    axis=1)

# Tri des rows selon, la moyenne puis selon le nombre de Nb NaN
dfDataFiltered_SE_SEC_ENRR = dfDataFiltered_SE_SEC_ENRR.sort_values(
    by=['Mean', 'Nb NaN'], ascending=[False, True])


# Tracé des classements de pays pour cette indicateur
fig, ax = plt.subplots(1, 2)
sns.barplot(x='Mean', y='Country Name',
            data=dfDataFiltered_SE_SEC_ENRR.nlargest(25, 'Mean'), ax=ax[0]).set_title('Top 25 Countries')
sns.barplot(x='Median', y='Country Name',
            data=dfDataFiltered_SE_SEC_ENRR.nlargest(25, 'Median'),  ax=ax[1]).set_title('Top 25 Countries')
fig.show()
plt.tight_layout()
```
